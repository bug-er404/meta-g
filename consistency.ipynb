# Minimap2 mappings

```python
# mapping reads to contigs
contig_to_reads = {}
for contig in (minimap2[1].unique()):
    contig_to_reads[contig]=minimap2[minimap2[1]==contig]
```

```python
# add lineage to mapped reads
for contig in contig_to_reads:
    contig_to_reads[contig] = contig_to_reads[contig].merge(reads,left_on=0,right_on='# contig')
```

```python
contig_list = []
contig_class = []
reads_maj_class = []

for contig in contig_to_reads:
    # record contig
    contig_list.append(contig)
    # record contig classification from CAT
    contig_class.append(contigs[contigs['# contig']==contig].iloc[0]['lineage'])
    # record majority reads class
    reads_maj_class.append(contig_to_reads[contig]['lineage'].value_counts())
    
```
